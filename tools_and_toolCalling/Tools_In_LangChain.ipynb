2. Custom Tools - a. Using @tool decorator

```python
from langchain_core.tools import tool
#Step1 - Creation any function definition
def addnums(a,b):
  """Adding two numbers"""
  return a+b
```

```python
#Step2 -Add type hints to the function you want to make a tool
def addnums(a:int,b:int)->int:
  """Adding two numbers"""
  return a+b
```

```python
#Step3 -  Adding a tool decorator

@tool
def addnums(a:int,b:int)->int:
  """Adding two numbers"""
  return a+b
```

```python
result = addnums.invoke({"a":3, "b":5})
```

```python
print(result)
```

```python
print(addnums.name)
print(addnums.description)
print(addnums.args)
```

2.Custom Tools - b. Using Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
from typing import Required
class MultiplyInput(BaseModel):
  a: int = Field(Required= True, description = "First number to add")
  b: int = Field(Required= True, description = "Second  number to add")
```

```python
def multiply_func(a:int,b:int)->int :
  """Multiply two numbers"""
  return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = "multiply",
    description = "Multiply two numbers",
    args_schema = MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3,'b':3})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

3. Custom Tools - c. Using BaseTool class

```python
from langchain.tools import  BaseTool
from typing import Type
from pydantic import BaseModel,Field
```

```python
#arg schema using pydantic
class MultiplyInput(BaseModel):
  a:int = Field(Required  = True, description = "First number to multiply")
  b:int = Field(Required  = True, description = "Second number to multiply")
```

```python
class MultiplyTool(BaseTool):
  name : str = "multiply"
  description : str = "Multiply two numbers"
  args_schema : Type[BaseModel] = MultiplyInput

  def _run(self,a:int,b:int)->int:
    return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3,'b':4})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args_schema)
```

##ToolKit

```python
from langchain_core.tools import tool

@tool
def add(a:int,b:int)->int:
   """Add two numbers"""
   return a+b

@tool
def multiply(a:int,b:int)->int:
  """Multiply two numbers"""
  return a*b
```

```python
class MathToolKit:
  def get_tools(self):
    return [add,multiply]
```

```python
toolkit = MathToolKit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
  print(tool.name, "=>",tool.description)
```
